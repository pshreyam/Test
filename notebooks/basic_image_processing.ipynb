```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(image, title="Image", cmap_type="gray"):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis("off")
```

```python
coffee_image = data.coffee()
coins_image = data.coins()
```

```python
coffee_image.shape
```

```python
coins_image.shape
```

```python
show_image(coins_image, "Coins Image")
```

```python
show_image(coffee_image, "Coffee Image")
```

```python
show_image(rocket_image, "Rocket Image")
```

```python
show_image(grayscale, "Grayscale Image")
```

```python
# Flip vertically using flipud function (upside down)
flipped_image = np.flipud(allianz_arena_image)
show_image(flipped_image)
```

```python
# Flip horizontally using fliplr function (left to right)
flipped_image = np.fliplr(allianz_arena_image)
show_image(flipped_image)
```
